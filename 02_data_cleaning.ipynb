```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\cleaned_healthcare_dataset.csv")
print("Initial Dataset Info:")
df.info()
df.columns = df.columns.str.strip()
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"], dayfirst=True)
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"], errors="coerce")
print("\nMissing values after verification:")
print(df.isnull().sum())
initial_rows = df.shape[0]
df = df.drop_duplicates()
final_rows = df.shape[0]
print(f"\nDuplicates removed: {initial_rows - final_rows}")
numeric_columns = ["Age", "Billing Amount", "Length of Stay"]
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")
print("\nFinal Dataset Info:")
df.info()
df.to_csv("healthcare_verified_cleaned.csv", index=False)
```
